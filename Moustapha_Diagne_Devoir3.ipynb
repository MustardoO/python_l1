## Méthodologie

**THÉORIE :**

La langue française comporte des mots de **nature** différente : noms, articles, pronoms, prépositions, conjonctions, adjectifs, adverbes, verbes...

Les index, quant à eux, permettent de **situer et répertorier** les endroits où sont présents les mots à travers un texte.
Ils sont par usage uniquement composés de mots pertinents. Par 'pertinents', on entend surtout les mots exprimant, par leur évocation, les idées générales qui caractérisent l'identité d'un texte.

Autrement dit, l'index doit permettre de **différencier les thématiques** d'un texte (et leur emplacement) parmi des milliards d'autres textes. Et les seuls mots capables de répondre à cette exigence sont les **noms (communs et propres), adjectifs et verbes**.

Ainsi, il faudrait idéalement que la `stoplist` soit composée **OU** de ce qu'on veut garder (shortlist) **OU** de ce qu'on ne veut pas garder (stoplist par définition).

En informatique, une `stoplist` est très pratique pour le référencement. Elle est garnie de `stopwords` ('mots vides' en français) considérés comme inutiles pour l'indexation.

Remarque : certains noms, verbes et adjectifs peuvent aussi être considérés comme `stopwords`.

## Affichage

La fonction `prd` est assez basique en l’état. Modifier cette fonction pour que la liste de lignes soit présentée sous la forme: `programme : 1-2, 7` quand il y a 2 numéros consécutifs ou plus; par exemple, 1, 2, 3, 5, la liste est présentée sous la
forme `1-3, 5`.

Pour ce faire, coder une fonction `presente` prenant une liste d'entier et retournant une chaîne de caractères au format voulu.

````python
# votre code ici

def presente(lst):
    
    lst = sorted(list(set(lst))) #on ordonne et supprime les doublons au cas où
        
    if lst[0] == lst[-1] : #on compare le premier et dernier élément, cas où il n'y a qu'un élément dans la liste
        
        
        return str(lst[0])
    
    
    else : #si la liste contient + de 1 élément
        
        
        chaine = "" #on initialise une chaîne de caractère
        
        
        for index, nombre in enumerate(lst) : #on itère dans la liste
            
            # AJOUTER tant qu'on est pas au dernier élément de la liste
            
            if chaine == "" : #si la chaine est vide
                
                if lst[index + 1] == nombre + 1 : #si l'élément suivant est dans la suite 
                
                    chaine += str(nombre) + '-'
                
                elif lst[index + 1] != nombre + 1 : #si l'élément suivant n'est pas dans la suite
                    
                    chaine += str(nombre) + ','
            
            #---------------------------------------------------------------------------------------------
            
            elif nombre == lst[-1] : #si on est arrivé au dernier élément
                
                chaine += str(lst[-1]) #on clôt la chaine en ajoutant le dernier élément
                
                break #on sort de la boucle maintenant pour éviter le cas chaine non vide qui suit
                
            #---------------------------------------------------------------------------------------------
            
            elif chaine != "" : #si la chaine n'est pas vide
                
                if chaine[-1] == '-' : #si termine par '-'
                
                    if lst[index + 1] == nombre + 1 : # si l'élément suivant est encore dans la suite
                        
                        continue
                        
                    elif lst[index + 1] != nombre + 1 : # si l'élément suivant n'est plus dans la suite
                        
                        chaine += str(nombre) + ','
                
                elif chaine[-1] == ',' : #si termine par ','
                    
                    if lst[index + 1] == nombre + 1 : #si l'élément suivant est dans la suite
                        
                        chaine += str(nombre) + '-'
                        
                    elif lst[index + 1] != nombre + 1 : #si l'élément suivant n'est pas dans la suite
                        
                        chaine += str(nombre) + ','
            
        return chaine
            
````

## Extraction des liens

A l'aide des indications vues dans le cours, écrire une fonction `extraire_liens(page)` retournant la liste des liens HTTP présents dans le contenu de la page. Pour rappel, la structure d'une balise contenant un lien est:

```
<a href="http..." ...>
```

````python
import re # On va utiliser les 'regular expressions'

def extraire_liens(page):
    # votre code ici
    
    lien = re.compile(r'"http.+?"') # note perso : r'<a href="http[\S]+"[\s\d\w"=<>]*</a>'
    
    #PROBLÈME : il faut purger les guillemets (") de lien.findall(page) sinon le robot ne pourra pas itérer
    
    lstLiensPurgés = []
    
    for lien in lien.findall(page) :  
        lstLiensPurgés.append(lien.strip(chr(34)))
    
    return lstLiensPurgés
````

## Nettoyage de la page

On souhaite nettoyer la page des différentes balises HTML qui s'y trouvent (une balise commence par le caractère `'<'` et termine par `'>'`). Vous devez réaliser une fonction `nettoie_page(page)` dont le rôle est de supprimer les balises HTML du contenu.

````python
def nettoie_page(page):
    # votre code ici
    
    # Aide sources : consulter https://www.w3schools.com/python/python_regex.asp, https://docs.python.org/3/library/re.html, https://jakevdp.github.io/WhirlwindTourOfPython/14-strings-and-regular-expressions.html, https://webdevdesigner.com/q/python-code-to-remove-html-tags-from-a-string-duplicate-66283/
    
    #pageSsVide = re.sub('\n|\t','',page) # pour une page sans retours chariot ni tabulation

    balise = re.compile(r'<[^>]+>') # ou '<.*?>' efficace mais -> #PROBLÈME : le '.' utilisé ci-dessous (hors []) dans les 're' exclus les retour chariot \n #SOLUTION : supprimer tous les retours chariots avant les balises
    
    texteDébalisé = re.sub(balise,'',page) # on cherche les balises dans la page et on les remplace par '' (supprime)
    
    #print(texteDébalisé)

    texteNettoyé = re.split("\s|\W",texteDébalisé) # on split() sur les espaces et les caractères NON alphanumériques

    
    #ETAPES restantes : supprimer les vides, les doublons + ne retenir que les mots non vide (dico français)
    
    texteNettoyé = list(set(list(filter(None,texteNettoyé)))) #on supprime les doublons, les vides
    
    #print(texteNettoyé)
    
    return sorted(texteNettoyé) #on renvoie la liste triée
````
